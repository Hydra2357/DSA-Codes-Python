```python
class MyThread(threading.Thread):
    def __init__(self, name, delay):
        super().__init__()
        self.name = name
        self.delay = delay

    def run(self):
        print(f"Thread {self.name} starting...")
        time.sleep(self.delay)
        print(f"Thread {self.name} finished after {self.delay} sec")

thread1 = MyThread("worker1", 1.5)
thread2 = MyThread("worker2", 2)

thread1.start()
thread2.start()

thread1.join()
thread2.join()
print("Main thread finished...")
```

```python
class MyThread(threading.Thread):
    def __init__(self, name, delay):
        super().__init__()
        self.name = name
        self.delay = delay

    def run(self):
        print(f"Starting {self.name} thread...")
        for i in range(2):
            print(f"{self.name}: Count {i}")
            time.sleep(self.delay)
            print(f"Finished {self.name}")

    threads = [
        MyThread("thread1", 1),
        MyThread("thread2", 2)
    ]
    print(type(threads))
    for thread in threads:
        thread.start()

    for thread in threads:
        thread.join()
```

```python
import threading
import time

def demon_worker():
    print(f"Thread {threading.current_thread().name} started")
    time.sleep(1)
    print("Demon thread finished...")

def normal_worker():
    print(f"Normal thread {threading.current_thread().name} started")
    time.sleep(2)
    print("Normal thread finished...")

d = threading.Thread(target=demon_worker, name="Daemon", daemon=True)
n = threading.Thread(target=normal_worker, name="Normal")

d.start()
n.start()

n.join()
print("THe main program exiting...")
```

Best Solutions to Stop Daemon Threads in Jupyter

```python
stop_event = threading.Event()

def daemon_worker1():
    print("daemon thread started...")
    while not stop_event.is_set():
        print("Daemon thread is running...")
        time.sleep(1)
        print("Daemon thread stopped gracefully...")

t1 = threading.Thread(target=daemon_worker1, daemon=True)
t1.start()
print("Thread started, do ur work")

stop_event.set()
print("Stop signal sent!")
```

```python
def long_running_task():
    print("Background task started...")
    while not stop_event.is_set():
        print("Working...")
        time.sleep(1)
    print("task stopped!!")

t = threading.Thread(target=long_running_task, daemon=True)
t.start()

stop_event.set()
t.join(timeout=2)
print("Thread should be stopped now")
```

```python
stop_event = threading.Event()

def background_job():
    print("Background job started...")
    i = 0
    while not stop_event.is_set():
        print(f"Running background job...count = {i}")
        i += 1
        time.sleep(1.2)
    print("Background job finished!")

job_thread = threading.Thread(target=background_job, daemon=True)
job_thread.start()

stop_event.set()
job_thread.join(timeout=5)
print("THread stopped!..")
```

```python
# Synchronization using lock
counter = 0
lock = threading.Lock()

def increment(name):
    global counter
    for _ in range(10000):
        with lock:
            counter += 1
    print(f"Thread {name} finished incrementing..")

t1 = threading.Thread(target=increment, args=("T1", ))
t2 = threading.Thread(target=increment, args=("T2", ))

t1.start()
t2.start()

t1.join()
t2.join()

print(f"Final counter: {counter}")
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

def task(n):
    time.sleep(1)
    return f"Task {n} completed"

with ThreadPoolExecutor(max_workers=5) as executor:
    futures = [executor.submit(task, i) for i in range(10)]

    for future in as_completed(futures):
        print(future.result())
```

#### Multiprocessing in Python

```python
from multiprocessing import Process
import os
import time

def worker(name):
    print(f"Process {name} - PID: {os.getpid()} - Starting...")
    time.sleep(2)
    print(f"Process {name} - PID: {os.getpid()} - Finished...")

if __name__ == "__main__":
    processes = []

    for i in range(3):
        p = Process(target=worker, args=f"Worker-{i}")
        processes.append(p)
        p.start()

    for p in processes:
        p.join()

    print("all processes completed!!")
```

```python
from multiprocessing import Process

class MyProcess(Process):
    def __init__(self, name):
        super().__init__()
        self.name = name

    def run(self):
        print(f"Custom Process {self.name} running - PID: {os.getpid()}")

if __name__ == "__main__":
    p1 = MyProcess("A")
    p2 = MyProcess("B")

    p1.start()
    p2.start()
    p1.join()
    p2.join()
```

```python
from multiprocessing import Process, Queue

def producer(q):
    for i in range(5):
        q.put(f"Item {i}")
    q.put(None)          # Sentinel

def consumer(q):
    while True:
        item = q.get()
        if item is None:
            break
        print("Consumed:", item)

if __name__ == "__main__":
    q = Queue()
    p1 = Process(target=producer, args=(q,))
    p2 = Process(target=consumer, args=(q,))

    p1.start()
    p2.start()
    p1.join()
    p2.join()
```
