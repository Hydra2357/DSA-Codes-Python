Hashing is the process of converting data (usually a key) into a fixed-size integer (the hash value) using a hash function.

In Python:

The built-in hash() function computes the hash of immutable objects.
Dictionaries (dict) and sets (set) use hash tables internally for O(1) average-case lookup, insert, and delete.

Only immutable objects can be hashed (str, int, float, tuple, frozenset, etc.).

```python
# Simplified demonstration of collision handling
class SimpleHashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [None] * size

    def _hash(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        original_index = index
        i = 1

        # Quadratic probing
        while self.table[index] is not None and self.table[index][0] != key:
            index = (original_index + i*i) % self.size
            i += 1
            if i > self.size:  # Table is full
                raise Exception("Hash table is full")

        self.table[index] = (key, value)

    def get(self, key):
        index = self._hash(key)
        original_index = index
        i = 1

        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (original_index + i*i) % self.size
            i += 1
            if i > self.size:
                break
        return None

# Test the implementation
ht = SimpleHashTable(5)
ht.insert("apple", 10)
ht.insert("banana", 20)
ht.insert("cherry", 30)
print(ht.get("banana"))  # 20
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __hash__(self):
        # Combine hash of attributes
        return hash((self.name, self.age))

    def __eq__(self, other):
        if not isinstance(other, Person):
            return False
        return self.name == other.name and self.age == other.age

"""Now Person objects can be dictionary keys"""
p1 = Person("Alice", 30)
p2 = Person("Alice", 30)
d = {p1: "Engineer"}
print(d[p2])
```

```python
import hashlib

def get_sha256(text):
    return hashlib.sha256(text.encode()).hexdigest()

print(get_sha256("hello"))
print(get_sha256("Hello"))  # Completely different
```

```python
class SimpleHashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]  # Separate Chaining

    def _hash(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        # Check if key already exists
        for pair in self.table[index]:
            if pair[0] == key:
                pair[1] = value
                return
        self.table[index].append([key, value])  # Chain

    def get(self, key):
        index = self._hash(key)
        for pair in self.table[index]:
            if pair[0] == key:
                return pair[1]
        return None
```

Summary
Hashing enables O(1) average lookup in dictionaries and sets

Collisions occur when different keys produce the same hash

Python uses open addressing with quadratic probing for collision resolution

Good hash functions distribute keys uniformly to minimize collisions

Hash randomization prevents hash flooding attacks

Always use immutable hashable types for dictionary keys
