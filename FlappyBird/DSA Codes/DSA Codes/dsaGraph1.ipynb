```python
#  Support for Weighted Graphs
class GraphWeighted:
    """supports unweighted graph (0, 1)"""
    def __init__(self, vertex, weighted=False):
        self.size = vertex
        self.weighted = weighted

        if weighted:
            self.mat = [[float('inf')]* vertex for _ in range(vertex)]

            for i in range(vertex):
                self.mat[i][i] = 0
        else:
            self.mat = [[0]*vertex for x in range(vertex)]

    def add_edge(self, source, destination, weight=1):
        if 0 <= source < self.size and 0 <= destination < self.size:
            if self.weighted:
                self.mat[source][destination] = weight
                self.mat[destination][source] = weight
            else:
                self.mat[source][destination] = 1
                self.mat[destination][source] = 1
        else:
            print("Invalid edge....")
```

```python
from collections import defaultdict
import heapq
from typing import Dict, List, Tuple, Optional, Set

class AdjListGraph:
    def __init__(self, directed: bool = False):
        """
        Initialize an empty graph using Adjacency List.
        directed=True  -> Directed Graph
        directed=False -> Undirected Graph
        """
        self.adj: Dict[int, List] = defaultdict(list)   # vertex -> list of neighbors
        self.directed = directed
        self.V = 0                                      # Number of vertices
        self.weights = {}                               # For easy weight lookup: ((u,v), weight)

    def add_vertex(self, vertex: int):
        """Add a new vertex to the graph"""
        if vertex not in self.adj:
            self.adj[vertex] = []
            self.V = max(self.V, vertex + 1)
        print(f"Vertex {vertex} added.")

    def add_edge(self, u: int, v: int, weight: int = 1):
        """Add an edge from u to v with optional weight"""
        # Add vertices if they don't exist
        self.add_vertex(u)
        self.add_vertex(v)

        # Add edge u -> v
        if v not in [n if isinstance(n, int) else n[0] for n in self.adj[u]]:
            if weight == 1:
                self.adj[u].append(v)
            else:
                self.adj[u].append((v, weight))
                self.weights[(u, v)] = weight

        # For undirected graph, add reverse edge
        if not self.directed:
            if u not in [n if isinstance(n, int) else n[0] for n in self.adj[v]]:
                if weight == 1:
                    self.adj[v].append(u)
                else:
                    self.adj[v].append((u, weight))
                    self.weights[(v, u)] = weight

        print(f"Edge added: {u} {'--' if not self.directed else '->'} {v} (weight={weight})")

    def remove_edge(self, u: int, v: int):
        """Remove edge between u and v"""
        if u in self.adj:
            # Remove v from u's list
            self.adj[u] = [x for x in self.adj[u]
                          if (x if isinstance(x, int) else x[0]) != v]

        if not self.directed and v in self.adj:
            self.adj[v] = [x for x in self.adj[v]
                          if (x if isinstance(x, int) else x[0]) != u]

        # Remove from weights
        self.weights.pop((u, v), None)
        if not self.directed:
            self.weights.pop((v, u), None)

        print(f"Edge removed: {u} - {v}")

    def remove_vertex(self, vertex: int):
        """Remove a vertex and all its connected edges"""
        if vertex in self.adj:
            del self.adj[vertex]

            # Remove all edges pointing to this vertex
            for u in list(self.adj.keys()):
                self.adj[u] = [x for x in self.adj[u]
                              if (x if isinstance(x, int) else x[0]) != vertex]

            # Clean weights
            self.weights = {k: w for k, w in self.weights.items()
                          if k[0] != vertex and k[1] != vertex}

            print(f"Vertex {vertex} and all its edges removed.")
            self.V -= 1

    def has_edge(self, u: int, v: int) -> bool:
        """Check if edge exists between u and v. O(degree) time"""
        if u not in self.adj:
            return False
        return any((x if isinstance(x, int) else x[0]) == v for x in self.adj[u])

    def get_neighbors(self, u: int) -> List:
        """Return list of neighbors of u"""
        if u not in self.adj:
            return []
        return self.adj[u]

    def get_weight(self, u: int, v: int) -> Optional[int]:
        """Get weight of edge u->v"""
        return self.weights.get((u, v))

    def degree(self, u: int) -> int:
        """Return degree (number of neighbors) of vertex u"""
        return len(self.adj.get(u, []))

    def get_all_vertices(self) -> List[int]:
        """Return all vertices"""
        return sorted(self.adj.keys())

    def print_graph(self):
        """Pretty print the adjacency list"""
        print("\n=== Adjacency List ===")
        for vertex in sorted(self.adj.keys()):
            neighbors = self.adj[vertex]
            print(f"{vertex} -> {neighbors}")
        print("=====================\n")

# Create Undirected Weighted Graph
g = AdjListGraph(directed=False)

g.add_edge(0, 1, 5)
g.add_edge(0, 2, 3)
g.add_edge(1, 3, 2)
g.add_edge(2, 3, 4)
g.add_edge(3, 4, 1)

g.print_graph()

print("Neighbors of 0:", g.get_neighbors(0))
print("Weight 0->2:", g.get_weight(0, 2))
print("Has edge 1-3?", g.has_edge(1, 3))
print("Degree of 3:", g.degree(3))

# Remove operations
g.remove_edge(0, 2)
g.remove_vertex(4)
g.print_graph()
```

```python
def bfs(self, start: int) -> List[int]:
    """Breadth First Search"""
    from collections import deque
    visited = set()
    queue = deque([start])
    visited.add(start)
    result = []

    while queue:
        u = queue.popleft()
        result.append(u)
        for nei in self.get_neighbors(u):
            v = nei if isinstance(nei, int) else nei[0]
            if v not in visited:
                visited.add(v)
                queue.append(v)
    return result

def dfs(self, start: int) -> List[int]:
    """Depth First Search"""
    visited = set()
    result = []

    def recurse(u):
        visited.add(u)
        result.append(u)
        for nei in self.get_neighbors(u):
            v = nei if isinstance(nei, int) else nei[0]
            if v not in visited:
                recurse(v)

    recurse(start)
    return result
```

```python
from collections import deque
from typing import List, Dict, Set, Any, Optional

def bfs_graph(graph: Dict[Any, List[Any]], start: Any) -> List[Any]:
    """
    BFS traversal of a graph represented as adjacency list.

    Args:
        graph: Dictionary where keys are nodes and values are lists of neighbors
        start: Starting node for BFS

    Returns:
        List of nodes in BFS order
    """
    # Set to keep track of visited nodes
    visited = set()

    # Queue for BFS - using deque for O(1) popleft operations
    queue = deque()

    # List to store the traversal order
    bfs_order = []

    # Start with the initial node
    visited.add(start)
    queue.append(start)

    while queue:
        # Remove the front node from queue
        current = queue.popleft()
        bfs_order.append(current)

        # Explore all neighbors of current node
        for neighbor in graph.get(current, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return bfs_order


def bfs_shortest_path(graph: Dict[Any, List[Any]], start: Any, end: Any) -> Optional[List[Any]]:
    """
    Find shortest path between start and end nodes using BFS.

    Args:
        graph: Graph as adjacency list
        start: Starting node
        end: Target node

    Returns:
        Shortest path as list of nodes, or None if no path exists
    """
    if start == end:
        return [start]

    visited = {start}
    queue = deque([(start, [start])])  # Store (node, path_to_node)

    while queue:
        current, path = queue.popleft()

        for neighbor in graph.get(current, []):
            if neighbor not in visited:
                if neighbor == end:
                    return path + [neighbor]
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))

    return None


def bfs_level_order(graph: Dict[Any, List[Any]], start: Any) -> Dict[int, List[Any]]:
    """
    BFS that returns nodes grouped by their distance level from start.

    Args:
        graph: Graph as adjacency list
        start: Starting node

    Returns:
        Dictionary mapping level number to list of nodes at that level
    """
    visited = {start}
    queue = deque([(start, 0)])  # (node, level)
    levels = {}

    while queue:
        current, level = queue.popleft()

        # Add current node to its level
        if level not in levels:
            levels[level] = []
        levels[level].append(current)

        for neighbor in graph.get(current, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, level + 1))

    return levels


def bfs_matrix(graph_matrix: List[List[int]], start: int) -> List[int]:
    """
    BFS for graph represented as adjacency matrix.

    Args:
        graph_matrix: 2D list where graph_matrix[i][j] = 1 if edge exists
        start: Starting node index

    Returns:
        List of nodes in BFS order
    """
    n = len(graph_matrix)  # Number of nodes
    visited = [False] * n
    bfs_order = []
    queue = deque()

    visited[start] = True
    queue.append(start)

    while queue:
        current = queue.popleft()
        bfs_order.append(current)

        # Check all possible neighbors
        for neighbor in range(n):
            if graph_matrix[current][neighbor] == 1 and not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)

    return bfs_order


# Example usage and test
if __name__ == "__main__":
    # Example graph as adjacency list
    graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D', 'E'],
        'C': ['A', 'F'],
        'D': ['B'],
        'E': ['B', 'F'],
        'F': ['C', 'E']
    }

    print("BFS Traversal from 'A':")
    result = bfs_graph(graph, 'A')
    print(f"Order: {result}")  # Output: ['A', 'B', 'C', 'D', 'E', 'F']

    print("\nShortest path from 'A' to 'F':")
    path = bfs_shortest_path(graph, 'A', 'F')
    print(f"Path: {path}")  # Output: ['A', 'C', 'F'] or ['A', 'B', 'E', 'F']

    print("\nNodes by level from 'A':")
    levels = bfs_level_order(graph, 'A')
    for level, nodes in levels.items():
        print(f"Level {level}: {nodes}")

    # Example with adjacency matrix
    matrix = [
        [0, 1, 1, 0, 0, 0],  # A
        [1, 0, 0, 1, 1, 0],  # B
        [1, 0, 0, 0, 0, 1],  # C
        [0, 1, 0, 0, 0, 0],  # D
        [0, 1, 0, 0, 0, 1],  # E
        [0, 0, 1, 0, 1, 0]   # F
    ]

    print("\nBFS on adjacency matrix (from node 0):")
    matrix_result = bfs_matrix(matrix, 0)
    print(f"Order: {matrix_result}")  # Output: [0, 1, 2, 3, 4, 5]
```
