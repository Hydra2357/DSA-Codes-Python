```python
def sliding_window_basic(arr, k):
    """Basic sliding window of size k"""
    res = []
    for i in range(len(arr) - k + 1):
        window = arr[i: i+k]
        res.append(window)
    return res

arr = [1 ,2, 3, 4, 5, 6, 7, 8,9]
print(sliding_window_basic(arr, 3))
print(sliding_window_basic(arr, 6))
print()
print(sliding_window_basic(arr, 4))
```

```python
from collections import deque

def sliding_window_deque(arr, k):
    """Efficient sliding window using deque"""
    res = []
    window = deque()

    for i, num in enumerate(arr):
        window.append(num)

        if len(window) > k:
            window.popleft()

        if len(window) == k:
            res.append(list(window))

    return res
arr = [1 ,2, 3, 4, 5, 6, 7, 8,9]
print(sliding_window_deque(arr, 3))
print(sliding_window_deque(arr, 6))
```

```python
def isMatch(arr1, arr2, start):
    n = len(arr1)
    for j in range(n):
        if arr1[(start + j) % n] != arr2[j]:
            return False
    return True


def minRotations(arr1, arr2):
    n = len(arr1)

    if n == 0 or len(arr2) != n:
        print("Invalid input: Arrays must be of same non-zero length")
        return

    leftRot = -1
    rightRot = -1

    for i in range(n):
        # Check for left rotation (rotating arr1 to the left by i positions)
        if leftRot == -1 and isMatch(arr1, arr2, i):
            leftRot = i

        # Check for right rotation (rotating arr1 to the right by i positions)
        # Right rotation by i == Left rotation by (n - i) % n
        if rightRot == -1 and isMatch(arr1, arr2, (n - i) % n):
            rightRot = i

    if leftRot == -1 and rightRot == -1:
        print("No rotation possible - Arrays are not rotations of each other")
        return

    print(f"Left Rotation: {leftRot if leftRot != -1 else 'Not possible'}")
    print(f"Right Rotation: {rightRot if rightRot != -1 else 'Not possible'}")

    min_rot = min(leftRot, rightRot) if leftRot != -1 and rightRot != -1 else \
              leftRot if leftRot != -1 else rightRot

    print(f"Minimum rotations needed: {min_rot}")
```

```python
# Maximum product of contiguous subarray
def maxProductFunction(nums):
    maxProduct = nums[0]
    minProduct = nums[0]
    ans = nums[0]

    for i in range(1, len(nums)):
        terms = (nums[i], nums[i]*maxProduct, nums[i]*minProduct)

        tempMax = max(terms)
        tempMin = min(terms)

        maxProduct = tempMax
        minProduct = tempMin

        ans = max(ans, maxProduct)

    return ans

print(maxProductFunction([2,5, 2, -2, 3, 2]))
print(maxProductFunction([ -2, 3, 2]))
print(maxProductFunction([2, 2, 2]))
```

```python
def mutiplyTheNonZeroNumbers(ipStr):
    nums = list(map(int, ipStr.split(',')))

    product = 1
    found = False

    for num in nums:
        if num != 0:
            product *= num
            found = True

    return product if found else 0

s = input("Enter the comma separate input:")
print("Product wothout including the Zeros: ", mutiplyTheNonZeroNumbers(s))

```

Longest substring no repeats	while s[right] in seen	expand → shrink\

At most K distinct chars	while len(chars) > k	expand → shrink\


```python
def fixed_sliding_window(arr, k):
    """
    Example: maximum sum of any subarray of size k.
    """
    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i - k]
        max_sum = max(max_sum, window_sum)

    return max_sum
```
