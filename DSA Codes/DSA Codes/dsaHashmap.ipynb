```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __hash__(self):
        return hash((self.x, self.y))      # Important!

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

p1 = Point(1, 2)
p2 = Point(1, 2)
d = {p1: "origin"}
print(d[p2])        # Works because __hash__ and __eq__ are defined
```

```python
# Demonstrating hash collisions impact (rare but possible)
import sys

# Python uses hash randomization (enabled by default)
# For strings, you might occasionally see collisions
class BadHash:
    def __hash__(self):
        return 42  # All instances hash to same value

    def __eq__(self, other):
        return id(self) == id(other)

# This will be O(n) due to collisions
bad_dict = {BadHash(): i for i in range(1000)}
# Lookup becomes O(n) for each key
```

```python
"""Group by Key"""
from collections import defaultdict

data = [("A", 10), ("B", 20), ("C", 30), ("D", 40), ("E", 50)]
groups = defaultdict(list)

for key, value in data:
    groups[key].append(value)

print(dict(groups))
```

```python
"""Safe nested access"""

def deep_get(d, keys, default=None):
    for key in keys:
        if isinstance(d, dict):
            d = d.get(key, default)
        else:
            return default
    return d
```

##### Hashset
Python’s built-in set is a Hash Set implementation — unordered collection of unique and hashable elements with average O(1) time complexity for insert, lookup, and delete.

```python
empty = set()
fruits = {"apple", "banana", "cherry"}
numbers = {1, 2, 3, 4, 2, 4, 4, 4, 2, 4}
numbers1 = {1, 2, 3, 4}
print(type(numbers))
print(type(numbers1))

print(fruits)
print(numbers)
```

```python
"""Removing the duplicates while preserving the order"""
def unique_preserve_order(arr):
    seen = set()
    result = []
    for x in arr:
        if x not in seen:
            seen.add(x)
            result.append(x)
    return result
```

```python
class Student:
    def __init__(self, id, name):
        self.id = id
        self.name = name

    def __hash__(self):
        return hash(self.id)

    def __eq__(self, other):
        return self.id == other.id

s = set()
s.add(Student(1, "A"))
s.add(Student(2, "B"))
s.add(Student(3, "C"))
```

```python
"""contains duplicate"""
def contains_duplicate(nums):
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False
```
