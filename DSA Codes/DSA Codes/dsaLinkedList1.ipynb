```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # 1. Insert at beginning
    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # 2. Insert at end
    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node

    # 3. Insert at specific position
    def insert_at(self, data, position):
        if position == 0:
            self.insert_beginning(data)
            return
        new_node = Node(data)
        curr = self.head
        for _ in range(position - 1):
            if not curr:
                print("Position out of range")
                return
            curr = curr.next
        if not curr:
            print("Position out of range")
            return
        new_node.next = curr.next
        curr.next = new_node

    # 4. Delete a node by value
    def delete(self, key):
        if not self.head:
            return
        # If head node is to be deleted
        if self.head.data == key:
            self.head = self.head.next
            return

        curr = self.head
        while curr.next and curr.next.data != key:
            curr = curr.next
        if curr.next:
            curr.next = curr.next.next
        else:
            print("Key not found")

    # 5. Print the list
    def print_list(self):
        curr = self.head
        while curr:
            print(curr.data, end=" -> ")
            curr = curr.next
        print("None")

    # 6. Length of list
    def length(self):
        count = 0
        curr = self.head
        while curr:
            count += 1
            curr = curr.next
        return count

    # 7. Search
    def search(self, key):
        curr = self.head
        position = 0
        while curr:
            if curr.data == key:
                return position
            curr = curr.next
            position += 1
        return -1

ll = SinglyLinkedList()
ll.insert_end(10)
ll.insert_end(20)
ll.insert_beginning(5)
ll.insert_at(15, 2)
ll.print_list()          # 5 -> 10 -> 15 -> 20 -> None
print("Length:", ll.length())
print("Position of 15:", ll.search(15))
ll.delete(15)
ll.print_list()
```

"""
TIME COMPLEXITY ANALYSIS

Operation                    | Time Complexity | Space Complexity
----------------------------|----------------|------------------
Insert at beginning         | O(1)           | O(1)
Insert at end              | O(n)           | O(1)
Insert at position         | O(n)           | O(1)
Delete from beginning      | O(1)           | O(1)
Delete from end            | O(n)           | O(1)
Delete by value            | O(n)           | O(1)
Search                     | O(n)           | O(1)
Access by index            | O(n)           | O(1)
Reverse                    | O(n)           | O(1)
Detect cycle               | O(n)           | O(1)
Find middle                | O(n)           | O(1)
Remove duplicates          | O(n)           | O(n)
Sort (Merge Sort)          | O(n log n)     | O(log n)
Merge sorted lists         | O(n + m)       | O(1)
Rotate                     | O(n)           | O(1)

ADVANTAGES OF LINKED LISTS:
1. Dynamic size
2. Efficient insertion/deletion at beginning
3. No memory wastage
4. Implementation of other data structures

DISADVANTAGES:
1. No random access
2. Extra memory for pointers
3. Not cache friendly
4. Reverse traversal not possible in singly linked list
"""

```python
class SinglyLinkedList:
    """Basic singly linked list with essential operations"""

    def __init__(self):
        self.head = None
        self._size = 0

    def __len__(self):
        """Return size of linked list"""
        return self._size

    def __str__(self):
        """String representation of linked list"""
        if not self.head:
            return "Empty List"

        current = self.head
        nodes = []
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " -> ".join(nodes) + " -> None"

    def is_empty(self):
        """Check if list is empty"""
        return self.head is None

    def insert_at_beginning(self, data):
        """Insert node at the beginning - O(1)"""
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        self._size += 1

    def insert_at_end(self, data):
        """Insert node at the end - O(n)"""
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            self._size += 1
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
        self._size += 1

    def insert_at_position(self, data, position):
        """Insert node at specific position - O(n)"""
        if position < 0 or position > self._size:
            raise IndexError("Position out of range")

        if position == 0:
            self.insert_at_beginning(data)
            return

        new_node = Node(data)
        current = self.head
        for _ in range(position - 1):
            current = current.next

        new_node.next = current.next
        current.next = new_node
        self._size += 1

    def delete_from_beginning(self):
        """Delete first node - O(1)"""
        if not self.head:
            raise IndexError("List is empty")

        deleted_data = self.head.data
        self.head = self.head.next
        self._size -= 1
        return deleted_data

    def delete_from_end(self):
        """Delete last node - O(n)"""
        if not self.head:
            raise IndexError("List is empty")

        if not self.head.next:  # Only one node
            deleted_data = self.head.data
            self.head = None
            self._size -= 1
            return deleted_data

        current = self.head
        while current.next.next:
            current = current.next

        deleted_data = current.next.data
        current.next = None
        self._size -= 1
        return deleted_data

    def delete_node(self, key):
        """Delete node by value - O(n)"""
        if not self.head:
            return False

        # If head node contains the key
        if self.head.data == key:
            self.head = self.head.next
            self._size -= 1
            return True

        current = self.head
        while current.next:
            if current.next.data == key:
                current.next = current.next.next
                self._size -= 1
                return True
            current = current.next

        return False

    def delete_at_position(self, position):
        """Delete node at specific position - O(n)"""
        if position < 0 or position >= self._size:
            raise IndexError("Position out of range")

        if position == 0:
            return self.delete_from_beginning()

        current = self.head
        for _ in range(position - 1):
            current = current.next

        deleted_data = current.next.data
        current.next = current.next.next
        self._size -= 1
        return deleted_data

    def search(self, key):
        """Search for a value - O(n)"""
        current = self.head
        position = 0
        while current:
            if current.data == key:
                return position
            current = current.next
            position += 1
        return -1

    def get_node_at_position(self, position):
        """Get node at specific position - O(n)"""
        if position < 0 or position >= self._size:
            raise IndexError("Position out of range")

        current = self.head
        for _ in range(position):
            current = current.next
        return current.data

    def reverse(self):
        """Reverse the linked list - O(n)"""
        prev = None
        current = self.head

        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

        self.head = prev
```

```python
class AdvancedLinkedList(SinglyLinkedList):
    """Extended linked list with advanced operations"""

    def __init__(self):
        super().__init__()

    def detect_cycle(self):
        """Detect cycle using Floyd's algorithm - O(n)"""
        slow = self.head
        fast = self.head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                return True
        return False

    def find_middle(self):
        """Find middle element using slow/fast pointers - O(n)"""
        if not self.head:
            return None

        slow = self.head
        fast = self.head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        return slow.data

    def create_cycle(self, position):
        """Create a cycle for testing (connects last node to node at position)"""
        if not self.head:
            return

        current = self.head
        cycle_node = None

        # Find the node at given position
        for i in range(position):
            if current:
                cycle_node = current
                current = current.next

        # Go to the last node
        while current.next:
            current = current.next

        # Connect last node to cycle_node
        if cycle_node:
            current.next = cycle_node

    def remove_duplicates(self):
        """Remove duplicate values - O(n) time, O(n) space"""
        if not self.head:
            return

        seen = set()
        current = self.head
        seen.add(current.data)

        while current.next:
            if current.next.data in seen:
                current.next = current.next.next
                self._size -= 1
            else:
                seen.add(current.next.data)
                current = current.next

    def remove_duplicates_sorted(self):
        """Remove duplicates from sorted list - O(n) time, O(1) space"""
        if not self.head:
            return

        current = self.head
        while current.next:
            if current.data == current.next.data:
                current.next = current.next.next
                self._size -= 1
            else:
                current = current.next

    def swap_nodes(self, x, y):
        """Swap two nodes without swapping data - O(n)"""
        if x == y:
            return

        # Search for x
        prev_x = None
        curr_x = self.head
        while curr_x and curr_x.data != x:
            prev_x = curr_x
            curr_x = curr_x.next

        # Search for y
        prev_y = None
        curr_y = self.head
        while curr_y and curr_y.data != y:
            prev_y = curr_y
            curr_y = curr_y.next

        if not curr_x or not curr_y:
            return

        # If x is not head
        if prev_x:
            prev_x.next = curr_y
        else:
            self.head = curr_y

        # If y is not head
        if prev_y:
            prev_y.next = curr_x
        else:
            self.head = curr_x

        # Swap next pointers
        curr_x.next, curr_y.next = curr_y.next, curr_x.next

    def pairwise_swap(self):
        """Swap elements pairwise - O(n)"""
        current = self.head

        while current and current.next:
            # Swap data
            current.data, current.next.data = current.next.data, current.data
            current = current.next.next

    def rotate(self, k):
        """Rotate list by k positions - O(n)"""
        if not self.head or k == 0:
            return

        # Find length and last node
        length = 1
        current = self.head
        while current.next:
            current = current.next
            length += 1

        # Adjust k
        k = k % length
        if k == 0:
            return

        # Connect last to head (make it circular)
        current.next = self.head

        # Find the node that will be new tail
        current = self.head
        for _ in range(length - k - 1):
            current = current.next

        # Update head and break the circle
        self.head = current.next
        current.next = None

    def merge_sorted(self, other_list):
        """Merge two sorted lists - O(n + m)"""
        merged = AdvancedLinkedList()

        curr1 = self.head
        curr2 = other_list.head

        # Dummy node to simplify logic
        dummy = Node(None)
        tail = dummy

        while curr1 and curr2:
            if curr1.data <= curr2.data:
                tail.next = curr1
                curr1 = curr1.next
            else:
                tail.next = curr2
                curr2 = curr2.next
            tail = tail.next

        # Attach remaining nodes
        if curr1:
            tail.next = curr1
        if curr2:
            tail.next = curr2

        merged.head = dummy.next
        merged._size = self._size + other_list._size
        return merged

    def sort(self):
        """Sort linked list using merge sort - O(n log n)"""
        if not self.head or not self.head.next:
            return

        self.head = self._merge_sort(self.head)

    def _merge_sort(self, head):
        """Helper function for merge sort"""
        if not head or not head.next:
            return head

        # Find middle and split
        middle = self._get_middle(head)
        next_to_middle = middle.next
        middle.next = None

        # Recursively sort both halves
        left = self._merge_sort(head)
        right = self._merge_sort(next_to_middle)

        # Merge sorted halves
        return self._sorted_merge(left, right)

    def _get_middle(self, head):
        """Get middle node for merge sort"""
        if not head:
            return head

        slow = head
        fast = head

        while fast.next and fast.next.next:
            slow = slow.next
            fast = fast.next.next

        return slow

    def _sorted_merge(self, left, right):
        """Merge two sorted lists"""
        if not left:
            return right
        if not right:
            return left

        if left.data <= right.data:
            result = left
            result.next = self._sorted_merge(left.next, right)
        else:
            result = right
            result.next = self._sorted_merge(left, right.next)

        return result

    def reverse_in_groups(self, k):
        """Reverse list in groups of k - O(n)"""
        if not self.head or k <= 1:
            return

        self.head = self._reverse_in_groups(self.head, k)

    def _reverse_in_groups(self, head, k):
        """Helper function to reverse in groups"""
        current = head
        next_node = None
        prev = None
        count = 0

        # Reverse first k nodes
        while current and count < k:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
            count += 1

        # Recursively reverse remaining
        if next_node:
            head.next = self._reverse_in_groups(next_node, k)

        return prev
```

```python
class PracticalLinkedList:
    """Real-world applications of linked lists"""

    @staticmethod
    def polynomial_representation():
        """Represent polynomial using linked list"""
        class PolyNode:
            def __init__(self, coeff, exp):
                self.coeff = coeff
                self.exp = exp
                self.next = None

        class Polynomial:
            def __init__(self):
                self.head = None

            def add_term(self, coeff, exp):
                new_node = PolyNode(coeff, exp)
                if not self.head:
                    self.head = new_node
                else:
                    current = self.head
                    while current.next:
                        current = current.next
                    current.next = new_node

            def display(self):
                if not self.head:
                    print("0")
                    return

                current = self.head
                terms = []
                while current:
                    if current.coeff != 0:
                        if current.exp == 0:
                            terms.append(str(current.coeff))
                        elif current.exp == 1:
                            terms.append(f"{current.coeff}x")
                        else:
                            terms.append(f"{current.coeff}x^{current.exp}")
                    current = current.next
                print(" + ".join(terms))

            def evaluate(self, x):
                result = 0
                current = self.head
                while current:
                    result += current.coeff * (x ** current.exp)
                    current = current.next
                return result

        # Test polynomial: 3x^2 + 2x + 1
        poly = Polynomial()
        poly.add_term(1, 0)
        poly.add_term(2, 1)
        poly.add_term(3, 2)

        print("Polynomial: ", end="")
        poly.display()
        print(f"Evaluation at x=2: {poly.evaluate(2)}")

    @staticmethod
    def music_playlist():
        """Simulate music playlist using linked list"""
        class Song:
            def __init__(self, title, artist):
                self.title = title
                self.artist = artist
                self.next = None
                self.prev = None  # For doubly linked list feature

        class Playlist:
            def __init__(self):
                self.head = None
                self.current = None

            def add_song(self, title, artist):
                new_song = Song(title, artist)
                if not self.head:
                    self.head = new_song
                else:
                    temp = self.head
                    while temp.next:
                        temp = temp.next
                    temp.next = new_song
                print(f"Added: {title} by {artist}")

            def display(self):
                if not self.head:
                    print("Playlist is empty")
                    return

                print("\n🎵 Playlist:")
                temp = self.head
                position = 1
                while temp:
                    print(f"{position}. {temp.title} - {temp.artist}")
                    temp = temp.next
                    position += 1

            def play_next(self):
                if self.current and self.current.next:
                    self.current = self.current.next
                    print(f"▶️ Now playing: {self.current.title}")
                elif self.head:
                    self.current = self.head
                    print(f"▶️ Now playing: {self.current.title}")

            def remove_song(self, title):
                if not self.head:
                    return False

                if self.head.title == title:
                    self.head = self.head.next
                    print(f"Removed: {title}")
                    return True

                current = self.head
                while current.next:
                    if current.next.title == title:
                        current.next = current.next.next
                        print(f"Removed: {title}")
                        return True
                    current = current.next

                print(f"Song '{title}' not found")
                return False

        # Test playlist
        playlist = Playlist()
        playlist.add_song("Shape of You", "Ed Sheeran")
        playlist.add_song("Blinding Lights", "The Weeknd")
        playlist.add_song("Dance Monkey", "Tones and I")
        playlist.display()

        playlist.play_next()
        playlist.play_next()
        playlist.remove_song("Blinding Lights")
        playlist.display()

    @staticmethod
    def undo_redo_system():
        """Implement undo/redo using linked list"""
        class ActionNode:
            def __init__(self, action):
                self.action = action
                self.next = None
                self.prev = None

        class UndoRedoManager:
            def __init__(self):
                self.current = None
                self.head = None

            def perform_action(self, action):
                new_node = ActionNode(action)

                if not self.head:
                    self.head = new_node
                    self.current = new_node
                else:
                    # Clear redo history
                    self.current.next = None
                    new_node.prev = self.current
                    self.current.next = new_node
                    self.current = new_node

                print(f"Performed: {action}")

            def undo(self):
                if self.current and self.current.prev:
                    self.current = self.current.prev
                    print(f"Undo: Reverted '{self.current.next.action if self.current else 'None'}'")
                    return self.current.action
                else:
                    print("Nothing to undo")
                    return None

            def redo(self):
                if self.current and self.current.next:
                    self.current = self.current.next
                    print(f"Redo: Restored '{self.current.action}'")
                    return self.current.action
                else:
                    print("Nothing to redo")
                    return None

            def show_history(self):
                if not self.head:
                    print("No history")
                    return

                print("\n📝 Action History:")
                temp = self.head
                index = 1
                while temp:
                    marker = " ← Current" if temp == self.current else ""
                    print(f"{index}. {temp.action}{marker}")
                    temp = temp.next
                    index += 1

        # Test undo/redo
        print("\nUndo/Redo System:")
        manager = UndoRedoManager()
        manager.perform_action("Type 'Hello'")
        manager.perform_action("Type ' World'")
        manager.perform_action("Delete ' World'")

        manager.show_history()
        manager.undo()
        manager.undo()
        manager.redo()
        manager.show_history()

# Test practical applications
print("\n" + "="*50)
print("PRACTICAL APPLICATIONS")
print("="*50)

print("\n1. Polynomial Representation:")
PracticalLinkedList.polynomial_representation()

print("\n2. Music Playlist:")
PracticalLinkedList.music_playlist()

print("\n3. Undo/Redo System:")
PracticalLinkedList.undo_redo_system()
```

```python
class IterableLinkedList(SinglyLinkedList):
    """Linked list with iteration support"""

    def __iter__(self):
        """Make linked list iterable"""
        self._iter_current = self.head
        return self

    def __next__(self):
        """Get next element"""
        if self._iter_current is None:
            raise StopIteration
        data = self._iter_current.data
        self._iter_current = self._iter_current.next
        return data

    def __getitem__(self, index):
        """Support indexing"""
        if index < 0:
            index = self._size + index

        if index < 0 or index >= self._size:
            raise IndexError("Index out of range")

        return self.get_node_at_position(index)

    def __contains__(self, item):
        """Support 'in' operator"""
        return self.search(item) != -1

    def to_list(self):
        """Convert linked list to Python list"""
        result = []
        current = self.head
        while current:
            result.append(current.data)
            current = current.next
        return result

    @classmethod
    def from_list(cls, lst):
        """Create linked list from Python list"""
        sll = cls()
        for item in lst:
            sll.insert_at_end(item)
        return sll

# Test iterable features
def test_iterable():
    print("\n" + "="*50)
    print("ITERABLE FEATURES TESTING")
    print("="*50)

    ill = IterableLinkedList.from_list([10, 20, 30, 40, 50])
    print(f"List: {ill}")

    # Iteration
    print("Iterating:", end=" ")
    for item in ill:
        print(item, end=" ")
    print()

    # Indexing
    print(f"Element at index 2: {ill[2]}")
    print(f"Element at index -1: {ill[-1]}")

    # Contains check
    print(f"Is 30 in list? {30 in ill}")
    print(f"Is 100 in list? {100 in ill}")

    # Convert to list
    print(f"As Python list: {ill.to_list()}")

test_iterable()
```

##### Singly Linked List

🔹 1. Node Creation

Logic:
Create a container that stores data and a pointer to the next node.

Algorithm:

Allocate memory for node
Assign data
Set next = NULL
🔹 2. Insert at Beginning

Logic:
New node becomes the first node; shift head forward.

Algorithm:

Create new_node
Set new_node.next = head
Update head = new_node

Time: O(1)

🔹 3. Insert at End

Logic:
Traverse to last node and attach new node.

Algorithm:

Create new_node
If head == NULL → head = new_node, STOP
Traverse until curr.next == NULL
Set curr.next = new_node

Time: O(n)

🔹 4. Insert at Position (0-based)

Logic:
Reach previous node, adjust links.

Algorithm:

If pos < 0 → invalid
If pos == 0 → insert at beginning
Traverse to (pos-1)th node
Set:
new_node.next = curr.next
curr.next = new_node

Time: O(n)

🔹 5. Traversal (Print List)

Logic:
Visit each node sequentially.

Algorithm:

Start from head
While curr != NULL:
Print curr.data
Move curr = curr.next

Time: O(n)

🔹 6. Length of List

Logic:
Count number of nodes.

Algorithm:

Initialize count = 0
Traverse list:
Increment count
Return count

Time: O(n)

🔹 7. Search Element

Logic:
Check each node until match found.

Algorithm:

Start from head
For each node:
If data == key → return index
If not found → return -1

Time: O(n)

🔹 8. Delete from Beginning

Logic:
Remove first node by shifting head.

Algorithm:

If head == NULL → underflow
Set head = head.next

Time: O(1)

🔹 9. Delete from End

Logic:
Remove last node using second-last node.

Algorithm:

If list empty → return
If only one node → head = NULL
Traverse to second-last node
Set curr.next = NULL

Time: O(n)

🔹 10. Delete by Value

Logic:
Find node and bypass it.

Algorithm:

If list empty → return
If head.data == key → delete head
Traverse while curr.next.data != key
Set curr.next = curr.next.next

Time: O(n)

🔹 11. Reverse (Iterative)

Logic:
Reverse links using 3 pointers.

Algorithm:

Initialize:
prev = NULL
curr = head
While curr != NULL:
next = curr.next
curr.next = prev
prev = curr
curr = next
Set head = prev

Time: O(n), Space: O(1)

🔹 12. Reverse (Recursive)

Logic:
Reverse from second node onward, fix first node later.

Algorithm:

Base case:
If node is NULL or single → return node
Recursively reverse rest
Fix:
node.next.next = node
node.next = NULL
Return new head
🔹 13. Cycle Detection (Floyd’s Algorithm)

Logic:
Fast pointer moves 2 steps, slow moves 1 → if cycle, they meet.

Algorithm:

Initialize slow = fast = head
While fast and fast.next:
slow = slow.next
fast = fast.next.next
If slow == fast → cycle exists
Return False

Time: O(n), Space: O(1)

🔹 14. Find Middle Element

Logic:
Fast pointer reaches end, slow pointer at middle.

Algorithm:

Initialize slow = fast = head
While fast and fast.next:
Move slow by 1
Move fast by 2
Return slow.data

Time: O(n)

🔹 15. Remove Duplicates (Sorted List)

Logic:
Adjacent duplicates exist → compare consecutive nodes.

Algorithm:

Start from head
While curr.next exists:
If curr.data == curr.next.data:
Skip duplicate
Else move forward

Time: O(n)

🔹 16. Rotate List by k

Logic:
Break list at (n-k) and attach tail to head.

Algorithm:

Find length n and last node
Compute k = k % n
If k == 0 → return
Traverse to (n-k-1) node
Set:
new_head = curr.next
curr.next = NULL
tail.next = head
Update head = new_head

Time: O(n)

```python
# Creating the Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Creating the singly linked list
class SinglyLinkedListNew:
    def __init__(self):
        self.head = None

# Inserting at beginning
    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

#insert at End
    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node

# Insert at Position
    def insert_at_position(self, data, position):
        if position < 0:
            print("Invalid Position...")
            return

        if position == 0:
            self.insert_beginning(data)
            return

        new_node = Node(data)
        curr = self.head

        for _ in range(position -1):
            if not curr:
                print("Position out of range...")
                return
            curr = curr.next

        if not curr:
            print("Position out of range...")
            return

        new_node.next = curr.next
        curr.next = new_node

# Traversal and utility methods
    def print_list(self):
        if not self.head:
            print("List is empty...")
            return
        curr = self.head
        while curr:
            print(curr.data, end= " -> ")
            curr = curr.next
        print("None")

    def length(self):
        count = 0
        curr = self.head
        while curr:
            count += 1
            curr = curr.next
        return count

    def search(self, key):
        curr = self.head
        index = 0
        while curr:
            if curr.data == key:
                return index
            curr = curr.next
            index += 1
        return -1

    def delete_beginning(self):
        if not self.head:
            return
        self.head = self.head.next

    def delete_end(self):
        if not self.head:
            return
        if not self.head.next:
            self.head = None
            return

        curr = self.head
        while curr.next.next:
            curr = curr.next
        curr.next = None

    def delete_value(self, key):
        if not self.head:
            return False

        if self.head.data == key:
            self.head = self.head.next
            return True

        curr = self.head
        while curr.next and curr.next.data != key:
            curr = curr.next

        if curr.next:
            curr.next = curr.next.next
            return True
        return False

    def reverse(self):
        prev = None
        curr = self.head

        while curr:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node
        self.head = prev

    def reverse_recursive(self):
        def helper(node):
            if not node or not node.next:
                self.head = node
                return node
            rest = helper(node.next)
            node.next.next = node
            node.next = None
            return rest

        helper(self.head)

    # Floyd’s Cycle Detection Algorithm
    def has_cycle(self):
        if not self.head:
            return False

        slow = fast = self.head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        if slow == fast:
            return True
        return False

# Fine the middle node in one pass
    def find_middle(self):
        if not self.head:
            return None

        slow = fast = self.head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow.data

# Remove Duplicates - For soted List
    def remove_duplicates(self):
        if not self.head:
            return
        curr = self.head

        while curr.next:
            if curr.data == curr.next.data:
                curr.next = curr.next.next
            else:
                curr = curr.next

# rotate list by K position
    def rotate(self, k):
        if not self.head or not self.head.next:
            return

        #length finding
        length = 1
        tail = 1
        tail = self.head
        while tail.next:
            tail = tail.next
            length += 1

        k = k % length
        if k == 0:
            return

        # Find the new tail
        curr = self.head
        for _ in range(length - k -1):
            curr = curr.next

        new_head = curr.next
        curr.next = None
        tail.next = self.head
        self.head = new_head

if __name__ == "__main__":
    l1 = SinglyLinkedListNew()

    print("\n--- INSERT OPERATIONS ---")
    l1.insert_beginning(10)
    l1.insert_beginning(5)
    l1.insert_end(20)
    l1.insert_end(30)
    l1.insert_at_position(15, 2)   # position index (0-based)

    print("List after insertions:")
    l1.print_list()

    print("\n--- SEARCH ---")
    print("Index of 15:", l1.search(15))
    print("Index of 100 (not present):", l1.search(100))

    print("\n--- LENGTH ---")
    print("Length:", l1.length())

    print("\n--- DELETE OPERATIONS ---")
    l1.delete_beginning()
    print("After delete_beginning:")
    l1.print_list()

    l1.delete_end()
    print("After delete_end:")
    l1.print_list()

    l1.delete_value(15)
    print("After delete_value(15):")
    l1.print_list()

    print("\n--- REVERSE ---")
    l1.reverse()
    print("After iterative reverse:")
    l1.print_list()

    l1.reverse_recursive()
    print("After recursive reverse (back to original):")
    l1.print_list()

    print("\n--- MIDDLE ELEMENT ---")
    print("Middle:", l1.find_middle())

    print("\n--- REMOVE DUPLICATES (Sorted Case) ---")
    l2 = SinglyLinkedListNew()
    l2.insert_end(1)
    l2.insert_end(1)
    l2.insert_end(2)
    l2.insert_end(2)
    l2.insert_end(3)

    print("Before removing duplicates:")
    l2.print_list()

    l2.remove_duplicates()
    print("After removing duplicates:")
    l2.print_list()

    print("\n--- ROTATE LIST ---")
    l3 = SinglyLinkedListNew()
    for i in range(1, 6):
        l3.insert_end(i)

    print("Original list:")
    l3.print_list()

    l3.rotate(2)
    print("After rotating by 2:")
    l3.print_list()

    print("\n--- CYCLE DETECTION ---")
    l4 = SinglyLinkedListNew()
    l4.insert_end(1)
    l4.insert_end(2)
    l4.insert_end(3)

    print("Cycle present:", l4.has_cycle())
```

```python
# Circular Linked List
class CircularNode:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        """Safe display of circular linked list"""
        if not self.head:
            print("List is Empty...")
            return

        current = self.head
        print(" -> ", end=" ")
        while True:
            print(current.data, end=" -> ")
            current = current.next
            if current == self.head:
                break
        print("back to head")

    def display_safe(self):
        if not self.head:
            print("List is EMpty...")
            return

        current = self.head
        nodes = []
        count = 0
        max_nodes = 100

        while current and count < max_nodes:
            nodes.append(str(current.data))
            current = current.next
            count += 1
            if current == self.head:
                break

        if count >= max_nodes:
            print("Warning: possible infinte loop found!")

        print(" -> ".join(nodes)+ "back to head" if nodes else "Empty list")

    def traversal(self):
        if not self.head:
            return []

        result = []
        current = self.head
        while True:
            result.append(current.data)
            current = current.next
            if current == self.head:
                break
        return  result

    def length(self):
        """Return number of nodes"""
        if not self.head:
            return 0

        count = 0
        current = self.head
        while True:
            count += 1
            current = current.next
            if current == self.head:
                break
        return count

    # ====================== INSERTIONS ======================
    def insert_at_beginning(self, data):
        """Insert node at the beginning"""
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            new_node.next = new_node  # points to itself
            return

        # Find last node
        last = self.head
        while last.next != self.head:
            last = last.next

        # Insert at beginning
        new_node.next = self.head
        last.next = new_node
        self.head = new_node

    def insert_at_end(self, data):
        """Insert node at the end"""
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            new_node.next = new_node
            return

        # Find last node
        last = self.head
        while last.next != self.head:
            last = last.next

        # Insert at end
        last.next = new_node
        new_node.next = self.head

    def insert_at_position(self, position, data):
        """Insert at given position (1-based indexing)"""
        if position < 1:
            print("Invalid position")
            return

        new_node = Node(data)

        # Position 1 (beginning)
        if position == 1:
            self.insert_at_beginning(data)
            return

        if not self.head:
            if position == 1:
                self.head = new_node
                new_node.next = new_node
            else:
                print("Position out of range")
            return

        current = self.head
        count = 1

        while count < position - 1 and current.next != self.head:
            current = current.next
            count += 1

        if count == position - 1:
            new_node.next = current.next
            current.next = new_node
        else:
            print("Position out of range")

    # ====================== DELETIONS ======================
    def delete_from_beginning(self):
        """Delete node from beginning"""
        if not self.head:
            print("List is empty")
            return None

        if self.head.next == self.head:  # Only one node
            data = self.head.data
            self.head = None
            return data

        # Find last node
        last = self.head
        while last.next != self.head:
            last = last.next

        data = self.head.data
        last.next = self.head.next
        self.head = self.head.next
        return data

    def delete_from_end(self):
        """Delete node from end"""
        if not self.head:
            print("List is empty")
            return None

        if self.head.next == self.head:  # Only one node
            data = self.head.data
            self.head = None
            return data

        # Find second last node
        current = self.head
        while current.next.next != self.head:
            current = current.next

        data = current.next.data
        current.next = self.head
        return data

    def delete_by_value(self, value):
        """Delete first occurrence of value"""
        if not self.head:
            print("List is empty")
            return False

        # If head is the value
        if self.head.data == value:
            self.delete_from_beginning()
            return True

        current = self.head
        while current.next != self.head:
            if current.next.data == value:
                current.next = current.next.next
                return True
            current = current.next

        print(f"Value {value} not found")
        return False

    # ====================== SEARCH ======================
    def search(self, value):
        """Search for a value, return position (1-based) or -1"""
        if not self.head:
            return -1

        current = self.head
        position = 1
        while True:
            if current.data == value:
                return position
            current = current.next
            position += 1
            if current == self.head:
                break
        return -1

    # ====================== REVERSE ======================
    def reverse(self):
        """Reverse the circular linked list"""
        if not self.head or self.head.next == self.head:
            return  # Empty or single node

        prev = None
        current = self.head
        next_node = None
        first = self.head

        # Break the circle temporarily
        while True:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
            if current == first:
                break

        # Fix the circle
        self.head.next = prev  # Original head now points to new last
        self.head = prev       # New head

    # ====================== DETECT CIRCULAR ======================
    def is_circular(self):
        """Detect if list is circular (Floyd's cycle detection)"""
        if not self.head:
            return False

        slow = self.head
        fast = self.head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False

    # ====================== SPLIT ======================
    def split_into_two(self):
        """Split circular list into two circular lists (even/odd positions)"""
        if not self.head or self.head.next == self.head:
            return self, CircularLinkedList()  # Original and empty

        list1 = CircularLinkedList()
        list2 = CircularLinkedList()

        slow = self.head
        fast = self.head.next

        while fast != self.head and fast.next != self.head:
            # Add to list1
            if not list1.head:
                list1.head = Node(slow.data)
                list1.head.next = list1.head
            else:
                temp = list1.head
                while temp.next != list1.head:
                    temp = temp.next
                temp.next = Node(slow.data)
                temp.next.next = list1.head

            # Add to list2
            if not list2.head:
                list2.head = Node(fast.data)
                list2.head.next = list2.head
            else:
                temp = list2.head
                while temp.next != list2.head:
                    temp = temp.next
                temp.next = Node(fast.data)
                temp.next.next = list2.head

            slow = fast.next
            fast = slow.next if slow.next != self.head else self.head

        # Handle remaining nodes if odd length
        if slow != self.head:
            # Add remaining slow to list1
            if not list1.head:
                list1.head = Node(slow.data)
                list1.head.next = list1.head
            else:
                temp = list1.head
                while temp.next != list1.head:
                    temp = temp.next
                temp.next = Node(slow.data)
                temp.next.next = list1.head

        return list1, list2

if __name__ == "__main__":
    cll = CircularLinkedList()

    print("=== Circular Linked List Operations Demo ===")

    # Insertions
    cll.insert_at_end(10)
    cll.insert_at_end(20)
    cll.insert_at_end(30)
    cll.insert_at_beginning(5)
    cll.insert_at_position(3, 25)

    print("After insertions:")
    cll.display()
    print("Length:", cll.length())

    # Search
    print("Position of 25:", cll.search(25))
    print("Position of 99:", cll.search(99))

    # Deletion
    cll.delete_from_beginning()
    cll.delete_from_end()
    cll.delete_by_value(25)
    print("After deletions:")
    cll.display()

    # Reverse
    cll.reverse()
    print("After reverse:")
    cll.display()

    # Check circular nature
    print("Is circular?", cll.is_circular())

    # Split
    list1, list2 = cll.split_into_two() if cll.length() > 0 else (CircularLinkedList(), CircularLinkedList())
    print("Split List 1:")
    list1.display()
    print("Split List 2:")
    list2.display()
```

```python
class CircularLinkedListC:
    def __init__(self):
        self.head = None
        self.tail = None

    # Helper method to check if list is empty
    def is_empty(self):
        return self.head is None

    # 1. TRAVERSAL
    def traverse(self):
        """Traverse and print all elements"""
        if self.is_empty():
            print("List is empty")
            return

        current = self.head
        while True:
            print(current.data, end=" -> ")
            current = current.next
            if current == self.head:
                break
        print("(back to head)")

    # 2. INSERT AT BEGINNING
    def insert_at_beginning(self, data):
        """Insert node at the beginning"""
        new_node = Node(data)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
            new_node.next = new_node  # Point to itself
        else:
            new_node.next = self.head
            self.head = new_node
            self.tail.next = self.head  # Tail points to new head

    # 3. INSERT AT END
    def insert_at_end(self, data):
        """Insert node at the end"""
        new_node = Node(data)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
            new_node.next = new_node
        else:
            new_node.next = self.head
            self.tail.next = new_node
            self.tail = new_node

    # 4. INSERT AT POSITION (0-indexed)
    def insert_at_position(self, data, position):
        """Insert node at specified position"""
        if position < 0:
            raise ValueError("Position cannot be negative")

        if position == 0:
            self.insert_at_beginning(data)
            return

        new_node = Node(data)
        current = self.head
        count = 0

        # Traverse to insert position
        while count < position - 1:
            current = current.next
            count += 1
            if current == self.head:
                raise IndexError("Position out of range")

        new_node.next = current.next
        current.next = new_node

        # Update tail if inserting after tail
        if new_node.next == self.head:
            self.tail = new_node

    # 5. DELETE FROM BEGINNING
    def delete_from_beginning(self):
        """Delete node from beginning"""
        if self.is_empty():
            print("List is empty, cannot delete")
            return None

        deleted_data = self.head.data

        if self.head == self.tail:  # Only one node
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.tail.next = self.head

        return deleted_data

    # 6. DELETE FROM END
    def delete_from_end(self):
        """Delete node from end"""
        if self.is_empty():
            print("List is empty, cannot delete")
            return None

        deleted_data = self.tail.data

        if self.head == self.tail:  # Only one node
            self.head = None
            self.tail = None
        else:
            # Find second last node
            current = self.head
            while current.next != self.tail:
                current = current.next
            current.next = self.head
            self.tail = current

        return deleted_data

    # 7. DELETE BY VALUE
    def delete_by_value(self, value):
        """Delete first occurrence of value"""
        if self.is_empty():
            print("List is empty")
            return False

        # Check if head node contains the value
        if self.head.data == value:
            self.delete_from_beginning()
            return True

        current = self.head
        while current.next != self.head:
            if current.next.data == value:
                # Node to delete is found
                if current.next == self.tail:  # If it's the tail
                    self.tail = current
                current.next = current.next.next
                return True
            current = current.next

        print(f"Value {value} not found")
        return False

    # 8. SEARCH
    def search(self, value):
        """Search for a value and return its position"""
        if self.is_empty():
            return -1

        current = self.head
        position = 0

        while True:
            if current.data == value:
                return position
            current = current.next
            position += 1
            if current == self.head:
                break

        return -1

    # 9. LENGTH
    def length(self):
        """Return the number of nodes"""
        if self.is_empty():
            return 0

        count = 0
        current = self.head

        while True:
            count += 1
            current = current.next
            if current == self.head:
                break

        return count

    # 10. REVERSE
    def reverse(self):
        """Reverse the circular linked list"""
        if self.is_empty() or self.head == self.tail:
            return  # No need to reverse

        prev = self.tail
        current = self.head
        next_node = None

        while True:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node

            if current == self.head:
                break

        # Swap head and tail
        self.head, self.tail = self.tail, self.head

    # 11. DETECT CIRCULAR NATURE
    @staticmethod
    def detect_circular(linked_list):
        """Detect if a linked list is circular"""
        if not linked_list or linked_list.head is None:
            return False

        slow = linked_list.head
        fast = linked_list.head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True

        return False

    # 12. SPLIT CIRCULAR LIST
    def split_list(self):
        """Split circular linked list into two halves"""
        if self.is_empty():
            return (None, None)

        if self.head == self.tail:
            # Single node lists
            list1 = CircularLinkedList()
            list2 = CircularLinkedList()
            list1.insert_at_end(self.head.data)
            return (list1, list2)

        # Find middle using slow and fast pointers
        slow = self.head
        fast = self.head

        while fast.next != self.head and fast.next.next != self.head:
            slow = slow.next
            fast = fast.next.next

        # Check if even or odd length
        if fast.next.next == self.head:
            fast = fast.next

        # Split the list
        head1 = self.head
        head2 = slow.next

        # Create first half
        list1 = CircularLinkedList()
        list2 = CircularLinkedList()

        # First half
        current = head1
        while True:
            list1.insert_at_end(current.data)
            current = current.next
            if current == head2:
                break

        # Second half
        current = head2
        while True:
            list2.insert_at_end(current.data)
            current = current.next
            if current == head1:
                break

        return (list1, list2)

    # 13. DISPLAY SAFELY
    def display_safely(self, max_iterations=100):
        """Display list with safety limit to prevent infinite loops"""
        if self.is_empty():
            print("List is empty")
            return

        current = self.head
        count = 0

        while count < max_iterations and current:
            print(current.data, end=" -> ")
            current = current.next
            count += 1
            if current == self.head:
                print("(back to head)")
                return

        print("... (possible infinite loop)")

    # 14. JOSEPHUS PROBLEM SOLUTION
    def josephus(self, step):
        """Solve Josephus problem - eliminate every k-th person"""
        if self.is_empty():
            return None

        current = self.head
        while self.length() > 1:
            # Move step-1 times
            for _ in range(step - 1):
                current = current.next

            # Remove the next node
            to_remove = current.next
            if to_remove == self.head:
                self.head = self.head.next
            if to_remove == self.tail:
                self.tail = current

            current.next = to_remove.next
            current = current.next

        return current.data
```

```python
def demonstrate_operations():
    print("=" * 60)
    print("CIRCULAR LINKED LIST DEMONSTRATION")
    print("=" * 60)

    # Create a circular linked list
    cll = CircularLinkedListC()

    # 1. Insert operations
    print("\n1. INSERT OPERATIONS")
    print("-" * 30)
    cll.insert_at_end(10)
    cll.insert_at_end(20)
    cll.insert_at_beginning(5)
    cll.insert_at_position(15, 2)
    print("After insertions (5, 10, 15, 20):")
    cll.traverse()

    # 2. Length
    print(f"\nLength: {cll.length()}")

    # 3. Search
    print("\n2. SEARCH OPERATION")
    print("-" * 30)
    print(f"Position of 15: {cll.search(15)}")
    print(f"Position of 100: {cll.search(100)}")

    # 4. Delete operations
    print("\n3. DELETE OPERATIONS")
    print("-" * 30)
    print(f"Deleted from beginning: {cll.delete_from_beginning()}")
    print(f"Deleted from end: {cll.delete_from_end()}")
    print("After deletions:")
    cll.traverse()

    # 5. Reverse
    print("\n4. REVERSE OPERATION")
    print("-" * 30)
    cll.insert_at_end(25)
    cll.insert_at_end(30)
    print("Before reverse:")
    cll.traverse()
    cll.reverse()
    print("After reverse:")
    cll.traverse()

    # 6. Delete by value
    print("\n5. DELETE BY VALUE")
    print("-" * 30)
    cll.delete_by_value(20)
    print("After deleting 20:")
    cll.traverse()

    # 7. Split list
    print("\n6. SPLIT CIRCULAR LIST")
    print("-" * 30)
    cll2 = CircularLinkedListC()
    for i in range(1, 7):
        cll2.insert_at_end(i * 10)
    print("Original list:")
    cll2.traverse()

    list1, list2 = cll2.split_list()
    print("First half:")
    list1.traverse()
    print("Second half:")
    list2.traverse()

    # 8. Josephus Problem
    print("\n7. JOSEPHUS PROBLEM")
    print("-" * 30)
    josephus_list = CircularLinkedListC()
    for i in range(1, 8):  # 7 people
        josephus_list.insert_at_end(i)
    print("Initial circle (7 people):")
    josephus_list.traverse()
    survivor = josephus_list.josephus(3)  # Eliminate every 3rd person
    print(f"Survivor (every 3rd eliminated): {survivor}")

    # 9. Detect circular nature
    print("\n8. DETECT CIRCULAR NATURE")
    print("-" * 30)
    normal_list = CircularLinkedListC()
    normal_list.insert_at_end(1)
    normal_list.insert_at_end(2)
    print(f"Is normal_list circular? {CircularLinkedListC.detect_circular(normal_list)}")

    # 10. Safe display
    print("\n9. SAFE DISPLAY")
    print("-" * 30)
    cll.display_safely()

def josephus_problem_example():
    """Detailed Josephus problem demonstration"""
    print("\n" + "="*60)
    print("JOSEPHUS PROBLEM DETAILED DEMONSTRATION")
    print("="*60)

    n = 41  # Number of people
    k = 3   # Eliminate every 3rd person

    circle = CircularLinkedListC()
    for i in range(1, n + 1):
        circle.insert_at_end(i)

    print(f"Starting with {n} people, eliminating every {k}rd person")
    survivor = circle.josephus(k)
    print(f"The survivor is person {survivor}")

def exception_handling_example():
    """Demonstrate error handling"""
    print("\n" + "="*60)
    print("EXCEPTION HANDLING DEMONSTRATION")
    print("="*60)

    cll = CircularLinkedListC()

    # Delete from empty list
    print("\nDeleting from empty list:")
    cll.delete_from_beginning()
    cll.delete_from_end()

    # Search in empty list
    print(f"\nSearch in empty list: {cll.search(10)}")

    # Insert at invalid position
    try:
        cll.insert_at_position(100, 5)
    except IndexError as e:
        print(f"Error handled: {e}")

    # Basic operations after empty
    cll.insert_at_end(1)
    cll.traverse()

if __name__ == "__main__":
    demonstrate_operations()
    josephus_problem_example()
    exception_handling_example()
```
