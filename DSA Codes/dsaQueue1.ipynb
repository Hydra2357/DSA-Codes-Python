```python
"""Efficient Custom queue using two stack (Amortized O(1))"""
class EfficientQueue:
    def __init__(self):
        self.in_stack = []    # For enqueue
        self.out_stack = []   # For dequeue

    def enqueue(self, item):
        """Add item to queue - O(1)"""
        self.in_stack.append(item)

    def dequeue(self):
        """Remove and return front item - Amortized O(1)"""
        if self.is_empty():
            raise IndexError("Dequeue from empty queue")

        if not self.out_stack:
            # Transfer all elements from in_stack to out_stack
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())

        return self.out_stack.pop()

    def peek(self):
        """Return front item without removing - Amortized O(1)"""
        if self.is_empty():
            raise IndexError("Peek from empty queue")

        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())

        return self.out_stack[-1]

    def is_empty(self):
        """Check if queue is empty - O(1)"""
        return len(self.in_stack) == 0 and len(self.out_stack) == 0

    def size(self):
        """Return number of items - O(1)"""
        return len(self.in_stack) + len(self.out_stack)

    def clear(self):
        """Remove all items - O(1)"""
        self.in_stack.clear()
        self.out_stack.clear()

# Usage
q = EfficientQueue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.dequeue())  # 1
q.enqueue(4)
print(q.dequeue())  # 2
print(q.dequeue())  # 3
print(q.dequeue())  # 4
```

```python
import heapq

class PriorityQueue:
    def __init__(self):
        self.heap = []
        self.counter = 0  # For stable sorting

    def enqueue(self, item, priority):
        """Add item with priority (lower number = higher priority)"""
        heapq.heappush(self.heap, (priority, self.counter, item))
        self.counter += 1

    def dequeue(self):
        """Remove and return highest priority item"""
        if self.is_empty():
            raise IndexError("Dequeue from empty priority queue")
        return heapq.heappop(self.heap)[2]

    def peek(self):
        """View highest priority item without removing"""
        if self.is_empty():
            raise IndexError("Peek from empty priority queue")
        return self.heap[0][2]

    def is_empty(self):
        return len(self.heap) == 0

    def size(self):
        return len(self.heap)

# Usage
pq = PriorityQueue()
pq.enqueue("low priority task", 3)
pq.enqueue("high priority task", 1)
pq.enqueue("medium priority task", 2)

print(pq.dequeue())  # high priority task
print(pq.dequeue())  # medium priority task
print(pq.dequeue())  # low priority task
```

#### Custom Deque implementation

```python
class CustomDeque:
    def __init__(self, maxlen=None):
        self.items = []
        self.maxlen = maxlen

    def append(self, item):
        if self.maxlen and len(self.items) >= self.maxlen:
            self.items.pop(0)
        self.items.append(item)

    def appendleft(self, item):
        if self.maxlen and len(self.items) >= self.maxlen:
            self.items.pop()
        self.items.insert(0, item)

    def pop(self):
        if not self.items:
            raise IndexError("pop from empty deque")
        return self.items.pop()

    def popleft(self):
        if not self.items:
            raise IndexError("pop from empty deque")
        return self.items.pop(0)

    def extend(self, iterable):
        for item in iterable:
            self.append(item)

    def extendleft(self, iterable):
        for item in iterable:
            self.appendleft(item)

    def rotate(self, steps=1):
        if not self.items:
            return
        steps = steps % len(self.items)
        if steps > 0:
            for _ in range(steps):
                self.appendleft(self.pop())
        elif steps < 0:
            for _ in range(-steps):
                self.append(self.popleft())

    def __len__(self):
        return len(self.items)

    def __getitem__(self, index):
        return self.items[index]

    def __repr__(self):
        return f"CustomDeque({self.items})"

# Example usage
cdq = CustomDeque(maxlen=5)
for i in range(1, 7):
    cdq.append(i)
print(cdq)  # CustomDeque([2, 3, 4, 5, 6])

cdq.appendleft(1)
print(cdq)  # CustomDeque([1, 2, 3, 4, 5])

cdq.rotate(1)
print(cdq)  # CustomDeque([5, 1, 2, 3, 4])
```

```python
# Application of deque
def is_palindrome(text):
    """Check is string is palindrome using deque"""
    dq = deque(text.lower())

    while len(dq) > 1:
        if dq.popleft() != dq.pop():
            return False
    return True

print(is_palindrome("Mahesh"))
print(is_palindrome("RAR"))
print(is_palindrome("tttt"))
print(is_palindrome("max"))
print(is_palindrome("A man "))
```

```python
def sliding_window_maximum(nums, k):
    """Find maximum in each sliding window od size k"""
    result = []
    dq = deque()

    for i, num in enumerate(nums):
        while dq and dq[0] <= i-k:
            dq.popleft()

        while dq and nums[dq[-1]] <= num:
            dq.pop()

        dq.append(i)

        if i >= k-1:
            result.append(nums[dq[0]])
    return result

# Test
nums = [1, 3, -1, -3, 5, 3, 6, 7]
k = 3
print(sliding_window_maximum(nums, k))  # [3, 3, 5, 5, 6, 7]
```

```python
class TaskScheduler:
    def __init__(self, max_history=10):
        self.tasks = deque()
        self.history = deque(maxlen=max_history)
        self.redo_stack = deque()

    def add_task(self, task):
        self.tasks.append(task)
        self.history.append(('add', task))
        self.redo_stack.clear()  # Clear redo on new action

    def complete_task(self):
        if self.tasks:
            task = self.tasks.popleft()
            self.history.append(('complete', task))
            self.redo_stack.clear()
            return task
        return None

    def undo(self):
        if self.history:
            action, task = self.history.pop()
            if action == 'add':
                self.tasks.pop()
                self.redo_stack.append(('add', task))
            elif action == 'complete':
                self.tasks.appendleft(task)
                self.redo_stack.append(('complete', task))
            return True
        return False

    def redo(self):
        if self.redo_stack:
            action, task = self.redo_stack.pop()
            if action == 'add':
                self.tasks.append(task)
                self.history.append(('add', task))
            elif action == 'complete':
                self.tasks.popleft()
                self.history.append(('complete', task))
            return True
        return False

    def show_tasks(self):
        return list(self.tasks)

# Test
scheduler = TaskScheduler()
scheduler.add_task("Task 1")
scheduler.add_task("Task 2")
print(scheduler.show_tasks())  # ['Task 1', 'Task 2']

scheduler.complete_task()
print(scheduler.show_tasks())  # ['Task 2']

scheduler.undo()
print(scheduler.show_tasks())  # ['Task 1', 'Task 2']

scheduler.redo()
print(scheduler.show_tasks())  # ['Task 2']
```

```python
from collections import deque
import time

def compare_performance():
    """Compare deque vs list for different operations"""

    # List vs Deque for left insertion
    lst = []
    dq = deque()

    # List left insertion (O(n))
    start = time.time()
    for i in range(10000):
        lst.insert(0, i)
    list_time = time.time() - start

    # Deque left insertion (O(1))
    start = time.time()
    for i in range(10000):
        dq.appendleft(i)
    deque_time = time.time() - start

    print(f"List left insertion: {list_time:.4f} seconds")
    print(f"Deque left insertion: {deque_time:.4f} seconds")
    print(f"Deque is {list_time/deque_time:.1f}x faster")

compare_performance()
# Output shows deque significantly faster for left-end operations
```

```python
class CircularQueue:
    def __init__(self, capacity: int):
        if capacity <= 0:
            raise ValueError("Capacity must be positive")

        self.capacity = capacity
        self.queue = [None] * capacity
        self.front = -1
        self.rear = -1
        self.count = 0                     # Tracks number of elements

    # ===================== ALL OPERATIONS =====================

    def is_empty(self) -> bool:
        """Check if queue is empty"""
        return self.count == 0

    def is_full(self) -> bool:
        """Check if queue is full"""
        return self.count == self.capacity

    def enqueue(self, item) -> bool:
        """Add item to the rear of the queue"""
        if self.is_full():
            print("Queue is Full! Overflow")
            return False

        if self.is_empty():
            self.front = 0

        self.rear = (self.rear + 1) % self.capacity
        self.queue[self.rear] = item
        self.count += 1
        return True

    def dequeue(self):
        """Remove and return item from the front"""
        if self.is_empty():
            print("Queue is Empty! Underflow")
            return None

        item = self.queue[self.front]
        self.queue[self.front] = None          # Optional: clean memory

        self.front = (self.front + 1) % self.capacity
        self.count -= 1

        # Reset pointers when queue becomes empty
        if self.count == 0:
            self.front = self.rear = -1

        return item

    def peek(self):
        """View front element without removing it"""
        if self.is_empty():
            print("Queue is Empty!")
            return None
        return self.queue[self.front]

    def rear_element(self):
        """View rear element without removing it"""
        if self.is_empty():
            print("Queue is Empty!")
            return None
        return self.queue[self.rear]

    def size(self) -> int:
        """Return current number of elements"""
        return self.count

    def clear(self):
        """Empty the queue"""
        self.queue = [None] * self.capacity
        self.front = self.rear = -1
        self.count = 0

    def display(self):
        """Display current state of queue"""
        if self.is_empty():
            print("Queue is Empty")
            return

        print("Queue:", end=" ")
        i = self.front
        for _ in range(self.count):
            print(self.queue[i], end=" ")
            i = (i + 1) % self.capacity
        print(f"  (Front: {self.front}, Rear: {self.rear}, Count: {self.count})")
```

```python
class CircularQueueNew:
    def __init__(self, capacity):
        """initialize the circular queue with given capacity"""
        self.capacity = capacity
        self.queue = [None]* capacity

        self.front = -1 # points to front element
        self.rear = -1 #points to rear element
        self.size = 0 # current size queue

    def is_empty_new(self):
        """Checking the queu is empty or not?"""
        return self.size == 0

    def is_full_new(self):
        """check the queue is full or not?"""
        return self.size == self.capacity

    def enqueue(self, item):
        """Add item to rear of queue"""
        if self.is_full_new():
            raise Exception("Queue Overflow!! cannot enqueue to full queue...")
        if self.is_empty_new():
            self.front = 0

        self.rear = (self.rear + 1)% self.capacity
        self.queue[self.rear] = item
        self.size +=1
        print(f"Enqueued: {item}")

    def dequeue(self):
        """Remove and return the element from the front of queue"""
        if self.is_empty_new():
            raise Exception("Queue Underflow!! Cannot dequeue from the Empty queue...")
        removed_item = self.queue[self.front]
        self.queue[self.front] = None
    #     Update the front position
        self.front = (self.front + 1) % self.capacity
        self.size -= 1

        if self.is_empty_new():
            self.front = -1
            self.rear = -1

        print(f"Dequeued: {removed_item}")
        return removed_item

    def get_front_new(self):
        """Get the front element without removing it"""
        if self.is_empty_new():
            raise Exception("Queue is empty!!")
        return self.queue[self.front]

    def get_rear_new(self):
        """Getting the rear element from queue without removing it"""
        if self.is_empty_new():
            raise Exception("Queue is Empty..")
        return self.queue[self.rear]

    def display(self):
        """Display all elements in the queue"""
        if self.is_empty_new():
            print("Queue is Empty!!!")
            return
        print("Queue elements from the front to rear")

        if self.front <= self.rear and not self.is_empty_new():
            for i in range(self.front, self.rear+1):
                print(self.queue[i], end=" ")
        else:
            for i in range(self.front, self.capacity):
                print(self.queue[i], end=" ")

            for i in range(0, self.rear+1):
                print(self.queue[i], end=" ")
        print()

    def get_size(self):
        '''Get current size of queue'''
        return self.size

    def get_capacity(self):
        """Get current capacity of the queue"""
        return self.capacity

    def clear(self):
        """Clear the entire Queue"""
        self.queue = [None]* self.capacity
        self.front = -1
        self.rear = -1
        self.size = 0
        print("The Queue cleared successfully!")

    def print_status(self):
        """print the detail status of Queue"""
        print("Queue status")
        print(f"The Capacity of Queue: {self.capacity}")
        print(f"Size of Queue: {self.size}")
        print(f"Front Index: {self.front}")
        print(f"Rear Index: {self.rear}")
        print(f"Is Empty?: {self.is_empty_new()}")
        print(f"is Full?: {self.is_full_new()}")
        print(f"Queue array: {self.queue}")

if __name__ == "__main__":
    print("="*10)
    cq = CircularQueueNew(10)
    cq.enqueue(34)
    cq.enqueue(54)
    cq.enqueue(61)
    cq.enqueue(61)
    cq.enqueue(61)
    cq.enqueue(61)
    cq.display()

    cq.enqueue(45)
    cq.enqueue(232)
    cq.enqueue(56)
    cq.enqueue(56)
    cq.display()
    print()
    cq.print_status()
    print()

    try:
        cq.enqueue(45)
    except Exception as e:
        print(f"Error: {e}")
```
