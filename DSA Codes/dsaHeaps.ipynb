#### What is a Heap?
A Heap is a special complete binary tree that satisfies the heap property:

Min-Heap: Every parent node is smaller than or equal to its children.
Max-Heap: Every parent node is larger than or equal to its children.

Heaps are commonly used to efficiently get the smallest or largest element quickly.

Python's Built-in Heap: heapq Module
Python provides the heapq module, which implements a Min-Heap on top of a regular list.

Function,Description,Time Complexity\
"heappush(heap, x)",Push element x onto the heap,O(log n)\
heappop(heap),Pop and return smallest element,O(log n)\
heapify(list),Convert a list into a heap in-place,O(n)\
"heapreplace(heap, x)",Pop smallest and push x,O(log n)\
"heappushpop(heap, x)",Push x then pop smallest,O(log n)\
"nlargest(k, iterable)",Return k largest elements,O(n log k)\
"nsmallest(k, iterable)",Return k smallest elements,O(n log k)\

```python
import heapq
tasks = []
heapq.heappush(tasks, (1, "Write the code",))
heapq.heappush(tasks, (3, "review the code",))
heapq.heappush(tasks, (2, "Fix the bug",))

while tasks:
    priority, task = heapq.heappop(tasks)
    print(f"Priority: {priority}: {task}")
```

#### Implementation of Heap Sort

```python
import heapq

def heap_sort(arr):
    heapq.heapify(arr)

    sorted_arr = []
    while arr:
        sorted_arr.append(heapq.heappop(arr))

    return sorted_arr

arr = [1,545 ,2 ,34 ,56 ,67, 78 ,89, 23 ,534, 34]
sorted_list = heap_sort(arr[:])
print(sorted_list)
```

```python
# Pure Heap Sort (Max-Heap)

def heapify(arr, n, i):
    """maintain max-heap propert"""
    largest = i
    left = 2*i + 1
    right = 2*i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left
    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    for i in range(n//2 -1, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

    return arr

arr = [12 ,32, 45 ,56 ,67, 78]
heap_sort(arr)
print(arr)
```
