```python
class NodeChai:
    """Linked list representation of tree"""
    def __init__(self, value):
        self.left = None
        self.right = None
        self.data = value

def pre_order_chai(root):
#     If the root is Null then no printing
    if root is not None:
        # Printing the data in following order
        print(root.data, end=" ")
        pre_order_chai(root.left)
        pre_order_chai(root.right)

def in_order_chai(root):
    if root is not None:
        in_order_chai(root.left)
        print(root.data, end=" ")
        in_order_chai(root.right)

def post_order_chai(root):
    if root is not None:
        post_order_chai(root.left)
        post_order_chai(root.right)
        print(root.data, end=" ")

root = NodeChai(1)

root.left = NodeChai(2)
root.right = NodeChai(3)

root.left.left = NodeChai(4)
root.left.right = NodeChai(5)

root.right.left = NodeChai(6)
root.right.right = NodeChai(7)

print("Preorder traversal: ")
pre_order_chai(root)
print()
print("In order traversal:")
in_order_chai(root)
print()
print("Postorder traversal:")
post_order_chai(root)
```

```python
class Node:
    """By Grok"""
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            pass

    def _insert_recursive(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursive(node.left, data)
        else:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursive(node.right, data)
```

```python
from collections import deque

"""Level order traversal"""
def level_order_traversal(root_new):
    if root_new is None:
        return
    q = deque([root_new])
    while q:
        current = q.popleft()
        print(current.data, end=" ")

        if current.left:
            q.append(current.left)
        if current.right:
            q.append(current.right)

level_order_traversal(root_new)
"""BFS traversal"""
```

```python
class NodeChaiBST:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def insert_in_bst(root_bst, value):
    # Checking the root is none
    if root_bst is not None:
        # Calling the node class constructor
        return NodeChaiBST(value)

    # root data is same as the given value
    if root_bst == value:
        return root_bst
    if root_bst > value:
        root_bst.left = insert_in_bst(root_bst.left, value)
    else:
        root_bst.right = insert_in_bst(root_bst.right, value)

```

A Binary Search Tree is a special type of binary tree that maintains the following property:
For any node:
All nodes in the left subtree have values less than the node’s value.
All nodes in the right subtree have values greater than the node’s value.
Both left and right subtrees are also BSTs.

This property allows O(log n) time complexity for Search, Insert, and Delete (on average).

```python
class TreeNode:
    def __init__(self, val = 0):
        self.val = val
        self.left = None
        self.right = None

"""Binary search tree"""
class BinarySearchTree:
    def __init__(self):
        self.root = None

    def search(self, val):
        """Return true if value exists, else return False"""
        return self._search_recursive(self.root, val)

    def _search_recursive(self, node, val):
        if not node:
            return False
        if node.val == val:
            return True
        elif val < node.val:
            return self._search_recursive(node.left, val)
        else:
            return self._search_recursive(node.right, val)

    def search_iterative(self, val):
        current = self.root
        while current:
            if current.val == val:
                return True
            elif val < current.val:
                current = current.left
            else:
                current = current.right
        return False

    """Insertion in BST"""
    def insert(self, val):
        """Insert a new value into the BST"""
        self.root = self._insert_recursive(self.root, val)

    def _insert_recursive(self, node, val):
        # Condition for empty tree
        if not node:
            return TreeNode(val)
        # If not empty tree means contains one node or more nodes

        if val < node.val:
            node.left = self._insert_recursive(node.left, val)
        elif val > node.val:
            node.right = self._insert_recursive(node.right, val)
        return node

    def insert_iterative(self, val):
        if not  self.root:
            self.root = TreeNode(val)
            return

        current = self.root
        while True:
            if val < current.val:
                if current.left is None:
                    current.left = TreeNode(val)
                    break
                current = current.left

            elif val > current.val:
                if current.right is None:
                    current.right = TreeNode(val)
                    break
                current = current.right
            else:
                break

    """Deletion in BST"""
    def delete(self, val):
        """Delete the value from the BST"""
        self.root  = self._delete_recursive(self.root, val)

    def _delete_recursive(self, node, val):
        if not node:
            return None

        if val < node.val:
            node.left = self._delete_recursive(node.left, val)
        elif val > node.val:
            node.right = self._delete_recursive(node.right, val)
        else:
            # case 1: leaf node
            if not node.left and not node.right:
                return None

            # case 2: Node with only one child
            if not node.left:
                return node.right
            if not node.right:
                return node.left

            #case3: node with two children and find inorder successor
            successor = self._min_value_node(node.right)
            node.val = successor.val
            successor = self._delete_recursive(node.right, successor.val)

        return node

    def _min_value_node(self, node):
        """find the node with min value"""
        current = node
        while current.left:
            current = current.left
        return current

    """"Helper function"""
    def inorder_traversal(self):
        """Inorder traversal gives sorted order in BST"""
        result = []
        self._inorder_helper(self.root, result)
        return result

    def _inorder_helper(self, node, result):
        if node:
            self._inorder_helper(node.left, result)
            result.append(node.val)
            self._inorder_helper(node.right, result)

    def print_tree(self):
        """Print the tree in sorted order"""
        print("inorder traversal: ", self.inorder_traversal())
```

```python
if __name__ == "__main__":
    bst = BinarySearchTree()

    values = [50, 30, 70, 20, 40, 60, 80, 35, 55]
    for val in values:
        bst.insert(val)

    print("BST created with values:", values)
    bst.print_tree()

    print("Search 40:", bst.search(40))      # True
    print("Search 100:", bst.search(100))    # False

    print("\n--- After deleting 20 (leaf node) ---")
    bst.delete(20)
    bst.print_tree()

    print("\n--- After deleting 30 (one child) ---")
    bst.delete(30)
    bst.print_tree()

    print("\n--- After deleting 50 (root with two children) ---")
    bst.delete(50)
    bst.print_tree()
```
