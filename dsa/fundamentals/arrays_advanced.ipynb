```python
def reverse_array(arr):
    left = 0
    right = len(arr) -1

    while left < right:
        arr[left], arr[right] = arr[right], arr[left],
        left += 1
        right -= 1
    return arr

arr = [23 ,12, 2, 4, 5, 56, 79, 89]
print("Array before Reverse: ")
print(arr)
reverse_array(arr)
print("Array after the reverse: ")
print(arr)
```

```python
def reverse_array_using_recursion(arrnew, left, right):
    if left >= right:
        return
    arrnew[left], arrnew[right] = arrnew[right], arrnew[left]
    reverse_array_using_recursion(arrnew, left+1, right-1)

arrnew = [12, 34, 45, 65, 76, 67]
print(arrnew)
reverse_array_using_recursion(arrnew, 0, len(arrnew)-1)
print(arrnew)
```
