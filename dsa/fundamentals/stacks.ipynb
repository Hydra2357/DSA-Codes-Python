```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        raise IndexError("pop from an empty stack")

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        raise IndexError("peek from an empty stack")

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def __str__(self):
        return str(self.items)

if __name__ == "__main__":
    s1 = Stack()
    s1.push(10)
    print(s1)
    s1.push(20)
    print(s1)
    s1.push(30)
    print(s1)
    s1.push(40)
    print(s1)
    s1.push(50)
    print(s1)
    print()

    s1.pop()
    print(s1)
    s1.pop()
    print(s1)
    print(s1.peek())
    print(s1.pop())
    print(s1.size())
```

```python
from collections import deque

class StackDeque:
    def __init__(self):
        self.items = deque()

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if self.items:
            return self.items.pop()
        raise IndexError("pop from an empty stack")

    def peek(self):
        return self.items[-1] if self.items else None

from queue import LifoQueue

stack = LifoQueue(maxsize=10)   # bounded stack

stack.put(1)
stack.put(2)

print(stack.get())  # 2 (blocks if empty, unless timeout used)
print(stack.qsize())
```

```python
class AdvancedStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []  # For O(1) min tracking

    def push(self, x):
        self.stack.append(x)
        if not self.min_stack or x <= self.min_stack[-1]:
            self.min_stack.append(x)

    def pop(self):
        if not self.stack:
            return None
        x = self.stack.pop()
        if x == self.min_stack[-1]:
            self.min_stack.pop()
        return x

    def get_min(self):
        return self.min_stack[-1] if self.min_stack else None
```

```python
# Stack implementation using insert function
class StackChai:
    """Initialize an empty stack"""
    def __init__(self):
        self.s = []

    def length(self):
        return len(self.s)

    def push(self, value):
        self.s.insert(0, value)

    def peek(self):
        if len(self.s) == 0:
            raise Exception("Stack is Empty")
        else:
            return self.s[0]

    def pop(self):
        if len(self.s) == 0:
            raise Exception("Stack is Empty")
        else:
            return self.s.pop(0)


s1 = StackChai()
s1.push(10)
s1.push(20)
s1.push(30)
s1.push(40)
print(s1.peek())
s1.pop()
s1.pop()
print(s1.peek())
```

```python
# Stack implementation using the append function
class NewStack:
    def __init__(self):
        self.elements = []

    def push(self, element):
        self.elements.append(element)
        print(f"Pushed on stack: {element}")

    def is_empty(self):
        """check the stack is empty or not"""
        return len(self.elements) == 0

    def clear(self):
        """Remove all elements from the stack"""
        self.elements.clear()
        print("Stack is cleared....")

    def size(self):
        """:return the no of elements from the stack"""
        return len(self.elements)


    def pop(self):
        """Remove and return the top element from the stack"""
        if self.is_empty():
            raise IndexError("cannot pop from the stack...")
        return self.elements.pop()

    def peek(self):
        """Display the top element of stack"""
        if self.is_empty():
            raise IndexError("Cannot peek form empty stack...")
        return self.elements[-1]

    def display(self):
        """Display all elements from the stack"""

        if self.is_empty():
            print("The stack is Empty...")
        else:
            print("Stack:", self.elements)

    def search(self, element):
        """Return the distance from the top (-1 based) if item exits, else -1"""
        if element in self.elements:
            return len(self.elements) - self.elements.index(element)
        return -1

    def stack_to_list(self):
        """"Convert the stack to list"""
        return self.elements.copy()

    def reverse(self):
        """Reverse the stack"""
        self.elements.reverse()
        print("the stack is reversed....")

    def get_min(self):
        """get the minimum element from the stack"""
        if self.is_empty():
            raise IndexError("Cannot get min element from empty stack")
        return min(self.elements)

    def get_max(self):
        """"Get the max element from the stack"""
        if self.is_empty():
            raise IndexError("Cannot getting the Max element from the stack")
        return max(self.elements)
```

```python
from collections import deque

class DequeStack:
    """Stack implementation using deque for better performance"""

    def __init__(self):
        self.stack = deque()

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("Cannot pop from empty stack")
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            raise IndexError("Cannot peek at empty stack")
        return self.stack[-1]

    def is_empty(self):
        return len(self.stack) == 0

    def size(self):
        return len(self.stack)

    def clear(self):
        self.stack.clear()

    def display(self):
        print("Stack (top to bottom):", list(reversed(self.stack)))

    def get_all(self):
        return list(self.stack)
```
